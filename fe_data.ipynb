```python
import pandas as pd

# Pfad zur Datei als String
datei_pfad = "C:/Users/fiton/Python Introduction/UCL_23_f.csv"

# CSV-Datei importieren
df = pd.read_csv(datei_pfad)

# Überflüssige Spalten entfernen (leere Spalten)
df = df.dropna(axis=1)
```

```python
# Wins per Match berechnen
WpM = df['Wins'] / df['Matches']
df['WpM'] = WpM
```

```python
# Tabelle nach Wins per Match sortieren
df_sorted = df.sort_values(by='WpM', ascending=False)
```

```python
# Die besten 5 Teams mit den meisten Titeln visualisieren
import matplotlib.pyplot as plt
df_top5_titles = df.sort_values(by='Titles', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.bar(df_top5_titles['Club'], df_top5_titles['Titles'])
plt.xlabel('Club')
plt.ylabel('Anzahl der Titel')
plt.title('Top 5 Teams mit den meisten Titeln')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Tore pro Spiel berechnen
df['Goals_per_Match'] = df['Goals_F'] / df['Matches']

# Die Top 10 Teams nach Tore pro Spiel sortieren und anzeigen
df_top10_goals = df.sort_values(by='Goals_per_Match', ascending=False).head(10)

# Balkendiagramm der Tore pro Spiel
plt.figure(figsize=(10, 6))
plt.bar(df_top10_goals['Club'], df_top10_goals['Goals_per_Match'])
plt.xlabel('Club')
plt.ylabel('Tore pro Spiel')
plt.title('Top 10 Teams mit den meisten Toren pro Spiel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
